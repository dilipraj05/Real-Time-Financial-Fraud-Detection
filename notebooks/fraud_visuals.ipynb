```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
```

```python
# Load predicted dataset
df = pd.read_csv("data/fraud_scored.csv")
```

```python
y_true = df['is_fraud']
y_pred = df['fraud_pred']

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix (Fraud Detection)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.xticks([0.5, 1.5], ['Not Fraud', 'Fraud'])
plt.yticks([0.5, 1.5], ['Not Fraud', 'Fraud'])
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='fraud_pred', data=df, palette='Set2')
plt.title('Predicted Fraud vs Non-Fraud')
plt.xlabel('Prediction (0 = Not Fraud, 1 = Fraud)')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='amount', hue='fraud_pred', bins=50, kde=True, palette='husl')
plt.title('Amount Distribution by Fraud Prediction')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.xlim(0, 1000)
plt.show()
```
